<h2>Choosing most 10 loved and hated chocolates</h2>

```python
cdj=[]
cdd=[]
for i in cd6:
    cdj.append( [i,cd[i].value_counts()['DESPAIR']])
    cdd.append( [i,cd[i].value_counts()['JOY']])
    
```

```python
cdj=pd.DataFrame(cdj,columns=['Type','Joy'])
cdd=pd.DataFrame(cdd,columns=['Type','Despair'])
cdj.sort_values(by=['Joy'],inplace=True,ascending=False)
cdd.sort_values(by=['Despair'],inplace=True,ascending=False)
```

```python
cdd=cdd.head(10)
cdd.sort_values(by=['Despair'],inplace=True)
cdd
```

```python
cdj=cdj.head(10)
cdj.sort_values(by=['Joy'],inplace=True)
cdj
```

<h3>Visualizing final results</h3>

```python
plt.figure(figsize=(10,7))
plt.barh(cdj['Type'],cdj['Joy'])
plt.title('Top 10 loved chocolates')
plt.show()
```

```python
plt.figure(figsize=(10,7))
plt.barh(cdd['Type'],cdd['Despair'])
plt.title('Top 10 hated chocolates')
plt.show()
```
